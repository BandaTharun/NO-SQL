```python
queries = [
    {"first_name": "Bernardo"},
    {"first_name": "Vitoria", "last_name": "Rings"},
    {"ip_address": "204.93.6.241", "last_name": "Ellwood"},
    {"first_name": "Zorina", "last_name": "Mannin", "email": "zmanning@booking.com"}
]

datasets = ["25%", "50%", "75%", "100%"]
num_iterations = 31
results = []

for i, query in enumerate(queries):
    for j, dataset in enumerate(datasets):
        collection = db[dataset]
        first_execution_time = None
        execution_times = []
        
        # First execution
        start_time = time.time()
        result = list(collection.find(query))
        end_time = time.time()
        first_execution_time = (end_time - start_time) * 1000  # Converting to milliseconds
        
        # Subsequent iterations
        for k in range(1, num_iterations):  # Starting from 1, skipping the first execution
            start_time = time.time()
            result = list(collection.find(query))
            end_time = time.time()
            execution_times.append((end_time - start_time) * 1000)  # Converting to milliseconds
        
        mean_time = sum(execution_times) / len(execution_times)
        results.append({"query": query, "dataset": dataset, "first_execution_time": first_execution_time, "mean_time": mean_time})
        
        print(f"Query {i+1} finished executing for collection {dataset}")
```

```python
# Defineing the queries to run

queries = [
    {"first_name": "Bernardo"},
    {"first_name": "Vitoria", "last_name": "Rings"},
    {"ip_address":  "204.93.6.241", "last_name":"Ellwood"},
    {"first_name": "Zorina", "last_name": "Mannin", "email":"zmanning@booking.com"}
    
]


# Defineing the dataset sizes
datasets = ["25%", "50%", "75%", "100%"]


# Defineing the number of iterations to run for each query

num_iterations = 31

# Creating a list to store the results
results = []

# Runing each query for each collection and record the execution times
# the time  are stored in the results 

for i, query in enumerate(queries):
    for j, dataset in enumerate(datasets):
        collection = db[dataset]
        first_execution_times = []
        execution_times = []
        for k in range(num_iterations):
            start_time = time.time()
            result = list(collection.find(query))
            end_time = time.time()
            if k == 0:
                first_execution_times.append((end_time - start_time) * 1000) # Converting to milliseconds
            else:
                execution_times.append((end_time - start_time) * 1000) # Converting  to milliseconds
        first_mean_time = sum(first_execution_times) / len(first_execution_times)
        mean_time = sum(execution_times) / len(execution_times)
        results.append({"query": query, "dataset": dataset, "first_mean_time": first_mean_time, "mean_time": mean_time})
        
        print(f"Query {i+1} finished executing for collection {dataset}")

```

```python
import time

datasets = [25, 50, 75, 100]

queries = [
    "SELECT * FROM data_{} WHERE first_name='Bernardo' ALLOW FILTERING",
    "SELECT * FROM data_{} WHERE first_name='Vitoria' AND last_name='Rings' ALLOW FILTERING",
    "SELECT * FROM data_{} WHERE ip_address='204.93.6.241' AND last_name='Ellwood' ALLOW FILTERING",
    "SELECT * FROM data_{} WHERE first_name='Zorina' AND last_name='Mannin' AND email='zmanning@booking.com' ALLOW FILTERING",
]

# Defining the number of iterations to run for each query
num_iterations = 31

# Creating a list to store the results
results = []

# Connect to the Cassandra cluster and execute the queries

for dataset_size in datasets:
    for i, query_template in enumerate(queries):
        query = query_template.format(dataset_size)
        first_execution_times = []
        execution_times = []
        for k in range(num_iterations):
            start_time = time.time()
            result = session.execute(query)
            end_time = time.time()
            if k == 0:
                first_execution_times.append((end_time - start_time) * 1000) # Convert to milliseconds
            else:
                execution_times.append((end_time - start_time) * 1000) # Convert to milliseconds
        first_mean_time = sum(first_execution_times) / len(first_execution_times)
   
        mean_time = sum(execution_times) / len(execution_times)
        
        results.append({"query": query, "dataset_size": dataset_size, "first_mean_time": first_mean_time, "mean_time": mean_time})
        print(f"Query {i+1} for dataset size {dataset_size} finished executing")


```
